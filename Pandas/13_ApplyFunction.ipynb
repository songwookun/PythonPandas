## 데이터에 함수 적용(apply)

```python
# 원본 데이터 다시 로드
df = pd.read_excel('score_with_missing_values.xlsx', index_col='지원번호')

# 함수 정의
def add_cm(height):
    return str(height) + 'cm'

# 한 번만 실행
df['키'] = df['키'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # NaN이 아닌지 확인
        return lang.capitalize() #첫 글자는 대문자, 나머지는 소문자로
    return lang
df['포지션'] = df['포지션'].apply(capitalize)
df
```
