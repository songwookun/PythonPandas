## 한글 폰트 설정

```python
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # macOS
matplotlib.rcParams['font.size'] = 15 # 글자 크기 변경
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 해결
```

```python
import matplotlib.font_manager as fm

# 사용 가능한 한글 폰트 확인
korean_fonts = [font.name for font in fm.fontManager.ttflist if 'Apple' in font.name or 'Malgun' in font.name or 'Nanum' in font.name or 'Gothic' in font.name]
print("사용 가능한 한글 폰트:")
for font in set(korean_fonts):
    print(f"- {font}")
```
