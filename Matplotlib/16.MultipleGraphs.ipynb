# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
            
#matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # macOS
matplotlib.rcParams['font.size'] = 15 # 글자 크기 변경
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 해결
```

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # 2 X 2에 해당하는 plot 들을 생성
fig.suptitle('여러개의 그래프 만들기')

# 첫 번째 그래프
axs[0,0].bar(df['이름'], df['득점'], label='득점') # 데이터 설정
axs[0,0].set_title('첫 번째 득점 수 그래프') # 제목
axs[0,0].legend() # 범례
axs[0,0].set(xlabel='이름', ylabel='득점 점수') # x, y축 label
axs[0,0].set_facecolor('lightyellow') # 전면 색
axs[0,0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0,1].plot(df['이름'], df['블록'], label='블록')
axs[0,1].set_title('두 번째 블록,리바운드 그래프') 
axs[0,1].plot(df['이름'], df['리바운드'], label='리바운드')
axs[0,1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])
axs[1,0].set_title('세 번째 학생별 키 그래프') 

# 네 번째 그래프
axs[1, 1].plot
```
