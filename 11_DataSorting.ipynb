# 11. 데이터 정렬

```python
import pandas as pd
df=pd.read_excel('score_with_missing_values.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('키') # 키 기준으로 오름차순 정렬
```

```python
df.sort_values('키', ascending=False) # 키 기준으로 내림차순 정렬
```

```python
df.sort_values(['리바운드', '득점']) #리바운드, 득점 점수 기준으로 오름차순
```

```python
df.sort_values(['리바운드', '득점'], ascending=False) #리바운드, 득점 점수 기준으로 내림차순
```

```python
df.sort_values(['리바운드','득점'], ascending=[True, False]) #리바운드는 오름차순으로, 득점은 내림차순으로 정렬
```
