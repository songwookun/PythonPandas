# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df=pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 불 값으로 확인 True/False
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] #filt를 역으로 적용 
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '리바운드'] #키가 185 이상인 학생들의 리바운드 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '득점','어시스트']] # 키가 185 이상인 학생들의 이름, 득점, 어시스트 데이터
```

## 다양한 조건

### & 그리고 조건

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 키가 185 이상인 북산고 학생 데이터
```

### | 또는 조건

```python
df.loc[(df['키'] > 185) | (df['어시스트'] >= 7)] #키가 185보다 크거나, 어시스트가 7보다 크거나 같은 데이터
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # '송'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 '태'가 들어가는 사람을 제외한 데이터
```

```python
langs = ['북산고', '능남고']
filt = df['학교'].isin(langs) # 학교가 북산고이거나 능남고인 데이터
df[filt]
```

```python
langs = ['pf','c']
filt = df['포지션'].str.lower().isin(langs) # 포지션 데이터 값을 소문자로 변경 후 지정한 소문자 데이터 출력
df[filt]
```

```python
filt = df['포지션'].str.contains('P') # 포지션 데이터 값 중 P 값이 들어있는 데이터
df[filt]
```
