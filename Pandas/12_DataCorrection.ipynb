# 12. 데이터 수정

```python
import pandas as pd
df=pd.read_excel('score_with_missing_values.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
df
```

```python
df['포지션'].str.lower() # 포지션 값 소문자로 변경
```

```python
df['포지션'] = df['포지션'].str.lower() 
df
```

```python
df['포지션'] = df['포지션'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' #학교 데이터 + 등학교 추가
df
```

## Column 추가

```python
df['총합'] = df['득점'] + df['리바운드'] + df['어시스트'] + df['스틸'] + df['블록']
df
```

```python
df['결과'] = 'Faill' # 결과  column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 100, '결과'] = 'Pass' # 총합이 100보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 column을 삭제 
```

```python
df.drop(columns=['득점', '리바운드', '스틸'])
```

## Row 삭제

```python
df.drop(index='4번') # 4번 학생 데이터 row를 삭제
```

```python
filt = df['득점'] < 50 # 득점이 50점 미만 학생 만 필터링
df[filt]
```
