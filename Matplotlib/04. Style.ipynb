# 4. 스타일(style)

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # macOS
matplotlib.rcParams['font.size'] = 15 # 글자 크기 변경
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 해결
```

```python
x = [1,2,3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.plot(x,y, linewidth = 5) # 선 굵기 조절
```

## 마커(marker)

```python
plt.plot(x,y, marker = 'o') # 동그라미 마커 추가
```

```python
plt.plot(x,y, marker = 'o', linestyle = 'None') # 동그라미 마커 추가 및 선 제거
```

```python
plt.plot(x, y, marker='v') # 삼각형 마커 추가
```

```python
plt.plot(x, y, marker='v', markersize=10) # 마커 사이즈 증가
```

```python
plt.plot(x, y, marker='X', markersize=10) # 마커 X 추가
```

### matplotllib.markers 검색으로 마커 찾아 사용 가능

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor = 'red') #마커 태두리 색상 변경
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor = 'red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor = 'red', markerfacecolor = 'yellow') # 마커 노란색으로 변경
```

## 선 스타일

```python
plt.plot(x, y, linestyle = ':') # 선 스타일 점선으로 변경
```

```python
plt.plot(x, y, linestyle = '--') # 굵고 긴 점선
```

```python
plt.plot(x, y, linestyle = '-.') # 선과 점 스타일 섞어서
```

```python
plt.plot(x, y, linestyle = '-') # 실선
```

```python
plt.plot(x, y, color='pink') # 선 색상 변경
```

```python
plt.plot(x, y, color='#ff0000') # 선 RGB로 색상 변경
```

### matplotllib.color 검색으로 원하는 색상 찾기 가능

```python
plt.plot(x, y, color='b') # Base 컬러 
```

```python
plt.plot(x, y, color='g') 
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

```python
plt.plot(x, y, color= 'g', marker='o', linestyle = 'None')
```

## 축약어
matplotlib markerfacecolor mfc 검색하여 확인 가능

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec= 'blue', ls=':')
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.3) #alpha : 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=150) # dots per inch, 확대
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
