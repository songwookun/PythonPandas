# 10. 누적 막대 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

#matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # macOS
matplotlib.rcParams['font.size'] = 15 # 글자 크기 변경
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자 깨지는 현상 해결
```

```python
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
plt.bar(df['이름'], df['득점']) # 파란색상 그래프
plt.bar(df['이름'], df['리바운드']) # 주황색 그래프
```

```python
plt.bar(df['이름'], df['득점'])
plt.bar(df['이름'], df['리바운드'], bottom= df['득점']) # 득점을 아래에 넣고 위에 리바운드 그래프를 올림
```

```python
plt.bar(df['이름'], df['득점'])
plt.bar(df['이름'], df['리바운드'], bottom = df['득점'])
plt.bar(df['이름'], df['블록'], bottom = df['득점'] + df['리바운드']) #가장 하단에 득점, 그 다음 리바운드, 최상단 블록 점수
```

```python
plt.bar(df['이름'], df['득점'], label='득점')
plt.bar(df['이름'], df['리바운드'], bottom = df['득점'], label='리바운드')
plt.bar(df['이름'], df['블록'], bottom = df['득점'] + df['리바운드'], label='블록')

plt.xticks(rotation=60)
plt.legend()
```
