# 4. 파일 저장 및 열기
DataFrame 객체를 exel, csv, txt등의 형태의 파일로 저장 및 열기

```python
import pandas as pd
data ={
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '포지션' : ['PF', 'C', 'PG', 'SG', 'SF', 'PF', 'C', 'PG'],
    '득점' : [70, 45, 37, 101, 49, 72, 68, 12],
    '리바운드' : [11, 10, 5, 9, 0, 19, 26, 5],
    '어시스트' : [10, 7, 7, 6, 17, 8, 14, 10],
    '스틸' : [2, 9, 4, 3, 12, 1, 2, 2],
    '블록' : [1, 13, 4, 1, 4, 19, 15, 3],
}
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df.index.name = '지원번호'
df
```

## 저장하기
### CSV 파일로 저장

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index = False)
```

### 텍스트 파일로 저장(.txt)

```python
df.to_csv('score.txt', sep='\t') # tab으로 구분된 텍스트 파일
```

### 엑셀 파일로 저장

```python
df.to_excel('score.xlsx')
```

***
## 열기

### csv파일 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 지정된 갯수 만큼의 row를 건너뜀
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1,3,5])# 1,3,5 Row는 제외(0부터 시작)
df
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 갯수 만큼의 row 만 가져옴
df
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=4) # 처음 2 row무시, 이후에 4 row를 가져옴
df
```

### 텍스트 파일 열기(.txt)

```python
df = pd.read_csv('score.txt', sep="\t")
df
```

```python
df = pd.read_csv('score.txt', sep="\t", index_col='지원번호')
df
```

```python
df = pd.read_csv('score.txt', sep="\t")
df.set_index("지원번호", inplace=True)
df
```

### 엑셀 파일 열기

```python
df = pd.read_excel('score.xlsx')
df
```

```python
df=pd.read_excel('score.xlsx', index_col='지원번호')
df
```
